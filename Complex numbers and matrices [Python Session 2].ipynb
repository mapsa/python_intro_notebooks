## Argand Diagram

```python
import matplotlib.pyplot as plt
import numpy as np
L = np.array([ 0. +6.j,  1. +7.j,  2. +8.j,  3. +9.j,  4.+10.j])

for x in L:
    plt.plot([0,x.real],[0,x.imag],'ro-',label='python')
```

### Ploting polar form

```python
for x in L:
    plt.polar([0,np.angle(x)],[0,abs(x)],marker='o')
```

# Matrices

```python
import numpy as np
```

Matrix multiplication

```python
A = np.matrix([[1,2],[3,4]])
B = np.matrix([[2,0],[1,2]])
print A * B
print A.dot(B)
```

```python
print A[:2,:1]
```

Matrix Transpose

```python
print A.transpose()
```

Matrix Inverse

```python
print np.linalg.inv(A)
```

```python
print np.linalg.eig(A)
```
